```python
import os

# Task 1: Write Sales Records to a File
# -----------------------------
sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.txt", "w") as file:
    for s in sales:
        file.write(str(s) + "\n")

print("Task 1: Sales data written to file.\n")
```

```python
# Task 2: Read File in Different Ways
# -----------------------------
print("Task 2: Reading File")

# read()
with open("sales_data.txt", "r") as file:
    print("Using read():")
    print(file.read())

# readline()
with open("sales_data.txt", "r") as file:
    print("Using readline():")
    print(file.readline())

# readlines()
with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    sales_list = [int(line.strip()) for line in lines]
    print("Using readlines():")
    print(sales_list)

print()
```

```python
# Task 3: Append New Sales
# -----------------------------
new_sales = [5000, 2500, 1700]

with open("sales_data.txt", "a") as file:
    for s in new_sales:
        file.write(str(s) + "\n")

print("Task 3: Appended new sales.\n")

with open("sales_data.txt", "r") as file:
    content = file.read()
    print("Updated file contents:")
    print(content)

with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    print("Total number of lines:", len(lines))

print()
```

```python
# Task 4: Generate Summary Report
# -----------------------------
with open("sales_data.txt", "r") as file:
    sales = [int(line.strip()) for line in file.readlines()]

total_sales = sum(sales)
highest_sale = max(sales)
lowest_sale = min(sales)
average_sale = total_sales / len(sales)

print("Task 4: Sales Summary")
print("Total Sales:", total_sales)
print("Highest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)
print("Average Sale:", average_sale)

print()
```

```python
# Task 5: Create Product Info File
# -----------------------------
print("Task 5: Enter 3 product details")

with open("products.txt", "w") as file:
    for i in range(3):
        name = input("Enter product name: ")
        price = input("Enter price: ")
        file.write(name + " " + price + "\n")

print("\nProduct List:")

with open("products.txt", "r") as file:
    for line in file:
        name, price = line.strip().split()
        print("Product:", name, "| Price:", price)

print()
```

```python
# Task 6: Read File Safely
# -----------------------------
filename = input("Enter filename to open: ")

if os.path.exists(filename):
    with open(filename, "r") as file:
        print(file.read())
else:
    print("File not found. Please check the filename.")

print()
```

```python
# Task 7: Mini Project – Export Discounted Prices
# -----------------------------
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

discount = float(input("Enter discount percentage: "))

with open("discount_report.txt", "w") as file:

    total = 0
    count = 0

    file.write("Product | Original Price | Discounted Price\n")

    for product, price in prices.items():
        discounted = price - (price * discount / 100)
        file.write(f"{product} | {price} | {discounted}\n")

        total += discounted
        count += 1

    avg_discount = total / count

    file.write("\nSummary\n")
    file.write("Total Items: " + str(count) + "\n")
    file.write("Average Discounted Price: " + str(avg_discount))

print("\nDiscount Report:\n")

with open("discount_report.txt", "r") as file:
    print(file.read())
```
